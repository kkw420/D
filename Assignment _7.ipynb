```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/forestfire.csv')
```

```python
df.head()
```

MATPLOTLIB.PYPLOT

```python
area_burned_per_month = df.groupby('month')['area'].mean()
plt.plot(area_burned_per_month.index, area_burned_per_month.values)
plt.xlabel('Month')
plt.ylabel('Average area burned')
plt.grid('show')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(df['area'], bins = 30, color = 'green')
plt.xlabel('Area')
plt.ylabel('Frequency')
plt.show()
```

```python
fires_per_month = df['month'].value_counts()
plt.bar(fires_per_month.index, fires_per_month.values, color = 'red')
plt.xlabel('Month')
plt.ylabel('Fires per month')
plt.show()
```

```python
df.boxplot(column='temp', by='month')
plt.xlabel('Month')
plt.ylabel('Temp')
plt.show()
```

```python
plt.hist(df['wind'], bins = 40, color = 'green')
plt.xlabel('Wind')
plt.ylabel('Frequency')
plt.show()
```

SEABORN


```python
df1 = df.drop(['month', 'day'], axis = 1)
```

```python
plt.figure(figsize = (9,9))
sns.heatmap(df1.corr(), annot = True, cmap = 'coolwarm')
plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.distplot(df['area'])
plt.show()
```

```python
forest_fire_day_wise = df['day'].value_counts().sort_index()
sns.barplot(forest_fire_day_wise)
plt.show()
```

```python
df2 = df[df['area']>0]
sns.scatterplot(x = 'area', y = 'temp', data = df2)
plt.grid('show')
```
